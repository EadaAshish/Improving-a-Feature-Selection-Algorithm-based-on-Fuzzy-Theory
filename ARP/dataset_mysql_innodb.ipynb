```python
import pandas as pd
```

```python
df = pd.read_csv('dataset_mysql_innodb.csv')
df
```

```python
df.drop(['id','Filename'], axis = 1, inplace=True)
df
```

```python
codes = {'M' : 1, 'B' : 0}
for i in df.columns:
    df[i] = df[i].replace(codes)
df
```

```python
df = df.rename({'AgingRelatedBugs':'class'}, axis=1)
df
```

```python
df[df['class'] == 0]
```

```python
maj_set = df[df['class'] == 0]
maj_set = maj_set.reset_index(drop=True)
maj_set
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
from matplotlib import pyplot as plt
```

```python
scalar = MinMaxScaler()
scalar.fit(maj_set)
new_set = scalar.transform(maj_set)
new_set
```

```python
km = KMeans(n_clusters=32)
y_pred = km.fit_predict(new_set)
y_pred
```

```python
maj_set['cluster'] = y_pred
maj_set
```

```python
maj_set[maj_set['cluster']==0]
```

```python
col_list = list(df.columns)
mean_set = pd.DataFrame(columns=col_list)
mean_set
```

```python
for i in range(32):
    mean_set.loc[len(mean_set.index)] =  maj_set[maj_set['cluster']==i].mean()
mean_set
```

```python
df = pd.concat([mean_set, df[df['class'] == 1]])
df = df.reset_index(drop=True)
df
```

```python
X = df.drop(['class'], axis=1)
X
```

```python
Y = df['class']
Y
```

# FISHER SCORE

```python
FScore = []

for m in  range(len(col_list)):
        mu = df[col_list[m]].mean()

        n_1 = df[df['class'] == 0].count()[1]
        mu_1 = df[col_list[m]][df['class'] == 0].mean()
        var_1 = df[col_list[m]][df['class'] == 0].var()

        n_2 = df[df['class'] == 1].count()[1]
        mu_2 = df[col_list[m]][df['class'] == 1].mean()
        var_2 = df[col_list[m]][df['class'] == 1].var()

        inter_class = n_1*(mu_1-mu)**2 + n_2*(mu_2-mu)**2
        intra_class = (n_1-1)*var_1 + (n_2-1)*var_2

        FScore.append(inter_class / intra_class)

FScore
```

```python
temp = pd.DataFrame(list(zip(col_list, FScore)), columns =['col', 'score']) 
temp
```

```python
temp_sort = temp.sort_values(['score'], ascending=False)
temp_sort[:10]
```

# MRMR 10

```python
from sklearn.feature_selection import f_regression

F = pd.Series(f_regression(X, Y)[0], index = X.columns)
corr = X.corr().abs().clip(.00001)
```

```python
selected = []
feat_score = []
not_selected = list(X.columns)
```

```python
K = 10

for i in range(K):
    
    # compute FCQ score for all the (currently) excluded features (this is Formula 2)
    score = F.loc[not_selected] / corr.loc[not_selected, selected].mean(axis = 1).fillna(.00001)
    
    # find best feature, add it to selected and remove it from not_selected
    best = score.index[score.argmax()]
    selected.append(best)
    feat_score.append(score[best])
    not_selected.remove(best)
```
